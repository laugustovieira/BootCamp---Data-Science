#Importar Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

#Carga de Dados


```python
dados = pd.read_csv("/content/A151346189_28_143_208.csv", encoding="ISO-8859-1", skiprows=3,
                  skipfooter = 12, sep=";", thousands = '.', decimal = ',', engine = "python")
```

#Desafio 01: Escolher um título mais descritivo, que passe a mensagem adequada ao gráfico de barras.

```python
eixo_x = "Unidade da Federação"
eixo_y = "2008/Ago"
dados = dados.sort_values(by=eixo_y, ascending=False)
ax = dados.plot(x = eixo_x, y = eixo_y, kind= "bar", figsize = [10,7])
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
titulo = "Custo do SUS por Unidade da Federação em {}".format(eixo_y)
plt.title(titulo)
```

#Desafio 02: Fazer a mesma análise para o mês mais recente

```python
eixo_x = "Unidade da Federação"
eixo_y = dados.columns[-2]
dados = dados.replace("-",0)
dados[eixo_y] = dados[eixo_y].astype("float64")
dados = dados.sort_values(by=eixo_y, ascending=False)
ax = dados.plot(x = eixo_x, y = eixo_y, kind= "bar", figsize = [10,7])
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
titulo = "Custo do SUS por Unidade da Federação em {}".format(eixo_y)
plt.title(titulo)
```
