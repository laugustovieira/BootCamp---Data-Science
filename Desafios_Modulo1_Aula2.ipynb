#Importar Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

#Carga de Dados


```python
dados = pd.read_csv("/content/drive/MyDrive/Dados_BootCamp/A005117189_28_143_208_2021_ATENDIMENTO.csv", encoding="ISO-8859-1", skiprows=3,
                  skipfooter = 12, sep=";", thousands = '.', decimal = ',', engine = "python")
```

#Tratar Dados

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0,"Unidade da Federação")
colunas_usaveis.remove("Total")
usaveis = dados[colunas_usaveis]
usaveis = usaveis.set_index("Unidade da Federação")
```

#Desafio 01 : Colocar a legenda separada

```python
ax = usaveis.T.plot(figsize=(12,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.legend(bbox_to_anchor=(1,1.1))
plt.title("Custo Total do Atendimento por Unidade da Federação")
```

#Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferencia

```python
estados_selecionados = ["35 São Paulo","31 Minas Gerais", "33 Rio de Janeiro", 
                        "32 Espírito Santo", "41 Paraná"]
ax = usaveis.T[estados_selecionados].plot(figsize=(12,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.legend(bbox_to_anchor=(1.2,0.7)) 
plt.title("Custo Total do Atendimento por Unidade da Federação")
```
