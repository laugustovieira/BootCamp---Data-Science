```python
import pandas as pd
```

Realizando a leitura do arquivo, colocando no formato brasil pois está em utf-8, pulando as 3 primeiras linhas, utilizando o separador ";"

```python
dados = pd.read_csv("/content/A151346189_28_143_208.csv", encoding = "ISO-8859-1", skiprows= 3, sep= ";",skipfooter=12,
                    engine = 'python', thousands= ".", decimal=",")
dados.head() #verificar o inicio da tabela
```

```python
dados.tail() #verificar o final da tabela
```

```python
dados.mean() #leitura da media, porém com valores elevados a 7
```

```python
pd.options.display.float_format = "{:.2f}".format # formatando em duas casas decimais
```

```python
dados.mean() # dados já formatados
```

```python
dados["2008/Ago"].mean() #media da coluna de agosta
```

```python
dados.plot(x = "Unidade da Federação", y="2008/Ago", kind = "bar", figsize=(10,7))
#plotando um gráfico das unidades da federação dos gastos do mês de agosto em barra com um tamanho 10,7
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x = "Unidade da Federação", y="2008/Ago", kind = "bar", figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}")) #formatar o eixo y
plt.title(" VAlor por Unidade da Federação")
```
