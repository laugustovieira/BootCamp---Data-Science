```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x = "Unidade da Federação", y="2008/Ago", kind = "bar", figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}")) #formatar o eixo y
plt.title(" VAlor por Unidade da Federação")
```

#Desafio 02: Fazer a mesma análise para o mês mais recente

```python
dados = dados.replace("-",0)
mes_mais_recente = dados.columns[-2] # pegar a ultima coluna
mes_mais_recente

dados["2020/Ago"] = dados["2020/Ago"].astype("float64") #converte coluna para float
dados = dados.sort_values(by=mes_mais_recente, ascending= False) # ordenação da informação

ax = dados.plot(x = "Unidade da Federação", y=mes_mais_recente, kind = "bar", figsize=(10,4))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Gasto do SUS por Unidade da Federação")
```

```python
colunas_usaveis = dados.mean().index.tolist() # importante usar o tolist para transformar 
                                              # em uma lista de python que é mutável. O index não é
colunas_usaveis.insert(0,"Unidade da Federação")
colunas_usaveis
```

```python
usaveis = dados[colunas_usaveis]
usaveis
```

```python
#criando um indice
usaveis = usaveis.set_index("Unidade da Federação")
```

```python
#buscar a linha deste indice
usaveis.loc["12 Acre"]
```

```python
#virou a tabela (Linas passa a ser colunas e colunas passam a ser linhas) e plota o gráfico
#ficou distorcido por causa que o total está sendo considerado
usaveis.T.plot(figsize=(12,6))
```

```python
#Retirar o total
usaveis = usaveis.drop("Total", axis=1) # 0coluna 1 linha
```

```python
usaveis.T.plot(figsize= (12,6))
```

#Desafio 01 : Colocar a legenda separada

```python

ax = usaveis.T.plot(figsize= (12,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.legend(bbox_to_anchor=(1,1.1)) 
plt.title("Custo Total do Atendimento por Unidade da Federação")
```

#Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferencia

```python
estados_selecionados = ["35 São Paulo","31 Minas Gerais", "33 Rio de Janeiro", 
                        "32 Espírito Santo", "41 Paraná"]
ax = usaveis.T[estados_selecionados].plot(figsize= (12,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.legend(bbox_to_anchor=(1.2,0.7)) 
plt.title("Custo Total do Atendimento por Unidade da Federação")
```
